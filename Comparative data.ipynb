```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 
file_paths = [
    "D:/WorkSpace/MSC_code/BM3D/results_BM3D.csv",
    "D:/WorkSpace/MSC_code/nlm-image-denoising/results_nlm.csv",
    "D:/WorkSpace/MSC_code/medianfilter/results_medianfilt.csv",
    "D:/WorkSpace/MSC_code/avefilter/results_avefilter.csv",
]

# 
labels = ["BM3D", "nlm", "medianfilt", "avefilter"]

plt.figure(figsize=(10, 6))

# 
for path, label in zip(file_paths, labels):
    data = pd.read_csv(path)
    sns.lineplot(data=data, x='Gaussian Noise', y='PSNR', marker='o', label=label)

plt.title('PSNR vs. Gaussian Noise')
plt.legend()
plt.show()

plt.figure(figsize=(10, 6))

# 
for path, label in zip(file_paths, labels):
    data = pd.read_csv(path)
    sns.lineplot(data=data, x='Gaussian Noise', y='SSIM', marker='o', label=label)

plt.title('SSIM vs. Gaussian Noise')
plt.legend()
plt.show()
```
