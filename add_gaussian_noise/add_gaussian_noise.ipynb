```python
import cv2
import numpy as np

def add_gaussian_noise(image, sigma):
    """
    向图像中添加高斯噪声
    :param image: 输入图像
    :param sigma: 噪声的标准偏差
    :return: 带有噪声的图像
    """
    row, col, ch = image.shape
    mean = 0
    gauss = np.random.normal(mean, sigma, (row, col, ch))
    gauss = gauss.reshape(row, col, ch)
    noisy = image + gauss
    # 确保值在0到255之间
    noisy = np.clip(noisy, 0, 255).astype(np.uint8)
    return noisy

def main():
    image_path = "D:/WorkSpace/MSC_code/add_gaussian_noise/Set12_Dataset/12.png"  # 替换为你的图片路径
    image = cv2.imread(image_path)

    # 从10到70为噪声等级
    for sigma in range(10, 71):
        noisy_image = add_gaussian_noise(image, sigma)
        output_filename = f"noisy_{sigma}_{image_path.split('/')[-1]}"
        cv2.imwrite(output_filename, noisy_image)
        print(f"Saved noisy image with sigma={sigma} as {output_filename}")

if __name__ == "__main__":
    main()





```
